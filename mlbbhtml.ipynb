**Read HTML into DataFrame**

```python
# Load library
import pandas as pd

# specify url
url = 'https://mobile-legends.fandom.com/wiki/List_of_heroes'

# by default read_html parse html and turn it from list to dataframe
df = pd.read_html(url, header=0)[0]

# drop column 'Hero', 'price'. axis=1 means columns, while axis=0 means rows
df = df.drop(df.columns[[0, 7]], axis=1)

# save to csv. r'' will prevent error when path are using relative path
df.to_csv(r'C:\Users\wis\Documents\GitHub\webscraping\mlbbhtml.csv', index=False)
```
