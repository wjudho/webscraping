## **Without Loop**

```python
from bs4 import BeautifulSoup
import pandas as pd

with open("cnbc.html", "r") as f:
    page = f.read()
soup = BeautifulSoup(page, 'html.parser')

tt = soup.select('.mCBkyc.y355M.ynAwRc.MBeuO.nDgy9d')
channels = [c.previous_sibling.text for c in tt]
print(channels[0])
times = [tm.next_sibling.next_sibling.next_sibling.text for tm in tt]
print(times[0]) 
titles = [t.text for t in tt]
print(titles[0])
subtitles = [ts.next_sibling.text for ts in tt]
print(subtitles[0])

nk = soup.select('a.WlydOe')
links = [nn.get('href') for nn in nk]
print(links[0])
```

## **With Loop**

```python
from bs4 import BeautifulSoup
import pandas as pd

with open("cnbc.html", "r") as f:
    page = f.read()
soup1 = BeautifulSoup(page, 'html.parser')

nw = soup1.select('.mCBkyc.y355M.ynAwRc.MBeuO.nDgy9d')
all_news = []
for n in nw:
    channel = n.previous_sibling.text
    time = n.next_sibling.next_sibling.next_sibling.text
    title = n.text
    subtitle = n.next_sibling.text
    
for n2 in soup1.select('a.WlydOe'):
    link = n2.get('href')
    all_news.append([channel, time, title, subtitle, link])

df = pd.DataFrame(all_news, columns=['Channels', 'Times', 'Titles', 'Subtitles', 'Links'])
df.to_csv('cnbc.csv', index=False)
df
```
